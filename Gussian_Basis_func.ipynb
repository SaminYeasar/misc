```python
import numpy as np
import matplotlib.pyplot as plt
```

## Gaussian Basis Function:
$\phi_k(X) = \exp^{(-\frac{(x-\mu_k)^2}{2\sigma^2})}$

```python
def gauss_basis(X,mu,sig):
    return plt.plot(X,np.exp( -(X-mu)**2/(2*sig**2) ))
```

```python
# for a fixed mu, sigma
X =  np.linspace(-1,1,100)
mu = 0.2
sig = 0.4
gauss_basis(X,mu,sig)
```

```python
# sigma controls width - activation radius
mu = 0.2
sig = np.linspace(0.1,0.5,10)
for val in sig:
    gauss_basis(X,mu,val)
```

```python
# mu controls position along x-axis
sig = 0.4
mu = np.linspace(0,0.5,10)
for val in mu:
    gauss_basis(X,val,sig)
```
